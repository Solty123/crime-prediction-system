# 7. VARIABLE TRANSFORMATION


```python
data['DISTRICT'] = data['DISTRICT'].apply(lambda x: x.upper())
```

```python
len(data['DISTRICT'].unique())
```

```python
encoder = LabelEncoder()
```

```python
data['STATE/UT'] = encoder.fit_transform(data['STATE/UT'])
data['DISTRICT'] = encoder.fit_transform(data['DISTRICT'])
```

```python
data.head(400)
```

# PREDICTIVE MODELING


### Splitting features and target 

```python
X = data.drop(columns=['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls'], axis=1)
Y = data[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']]
```

```python
X.head
```

```python
Y.head
```

# Splitting the Data into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

### Linear Regression 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

### K-nearest neighbour

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)
```

```python
knn.fit(X_train, Y_train)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(X_train, Y_train)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, Y_train)
```

# EVALUATION 

### Linear Regression 

```python
# prediction on training data
training_data_prediction = lr.predict(X_train)

# R squared value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R squared value =', r2_train)
```

```python
# prediction on test data
test_data_prediction = lr.predict(X_test)

# R squared value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R squared value =', r2_test)
```

### K-Nearest Neighbour

```python
# prediction on training data
training_data_prediction = knn.predict(X_train)

# R squared value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R squared value =', r2_train)
```

```python
# prediction on test data
test_data_prediction = knn.predict(X_test)

# R squared value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R squared value =', r2_test)
```

### Decision Tree Regressor 

```python
# prediction on training data
training_data_prediction = dtr.predict(X_train)

# R squared value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R squared value =', r2_train)
```

```python
# prediction on test data
test_data_prediction = dtr.predict(X_test)

# R squared value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R squared value =', r2_test)
```

### Random Forest Regressor

```python
# prediction on training data
training_data_prediction = rfr.predict(X_train)

# R squared value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R squared value =', r2_train)
```

```python
# prediction on test data
test_data_prediction = rfr.predict(X_test)

# R squared value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R squared value =', r2_test)
```
